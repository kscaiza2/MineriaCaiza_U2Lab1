## Arboles de decision

```python
'''Importamos las respectivas librerias para el 
desarrollo del laboratorio y de igual manera con las librerias para la matriz
de confusion, arbol de decision y demas modulos'''
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
import pydotplus
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np
from sklearn.tree import export_graphviz
from six import StringIO 
from IPython.display import Image  
```

```python
'''Cargamos el conjunto de datos que hace enfasis en pacientes con diabetes'''
df = pd.read_csv("diabetes.csv")
'''Imprimimos los primeros 5 datos los cuales inician desde el 0'''
df.head()
```

```python
'''Verificamos si las variables fueron cargadas correctamente a traves
de la funcion info, comprobando que existen un total de  768  registros'''
df.info()
```

```python
'''Analisis de columnas de datos del dataset'''
df.keys()
```

```python
'''Verificamos nuestros resultados de salida'''
print("Resultados de salida:\n", df.Outcome)
```

```python
'''Separamos la variable a predecir de las variables predictoras, haciendo uso
de una variable feature columns que permite indicar la ubicacion
de cada una de estas'''
feature_cols = ['Pregnancies', 'Insulin', 'BMI', 'Age', 'Glucose', 
                'BloodPressure', 'DiabetesPedigreeFunction']
X = df[feature_cols] 
y = df.Outcome
```

```python
'''Usamos la funcion importada de sklearn train_test_split, el cual 
dividira el conjunto de datos de entrenamiento y prueba 
seleccionando registros al azar'''
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
'''Mediante la funcion DecisionTreeClassifier generamos la estrucuta del 
arbol de decision y a traves de max_depth se puede agregar la cantidad 
que debe tener el arbol'''
clf = DecisionTreeClassifier()
'''A traves dela funcion fit pasaremos como pametros las variables predictoras
como la variable a predecir'''
clf = clf.fit(X_train,y_train)
'''Calculo de la prediccion con los datos obtenidos'''
y_pred = clf.predict(X_test)
```

```python
'''Evaluacion del modelo: Se obtendra el acurracy, precision y recall
del modelo a traves de Metrics perteneciente a sklearn.'''
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

```python
'''De manera mas resumida se obtendran los datos
a traves de ka funcion classification report'''
print(metrics.classification_report(y_test, y_pred, digits=3))
```

## Visualizing the tree

We will plot the tree to see its root, branches, and nodes using the following functions:

* **StringIO()**: creates an object (empty in this case) to receive a string buffer (the tree will be created first as a string before as an image) in DOT (graph description language) format.
* **export_graphviz()**: this function exports the tree in DOT format, generating a representation of the decision tree, which is written into the ‘out_file’.
* **graph_from_dot_data()**: will use the DOT object to create the graph.
* **Image()**: uses raw data to return an image (png, jpeg, gif).

```python
'''A traves de la herramientas de pydotplus crearemos el grafico del arbol 
indicando la funcion a graficar y se le asignara un nombre respectivamente'''
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,
                feature_names = feature_cols,
                class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```

## Pruning

The complete decision tree is difficult to interpret due to the complexity of the outcome. Pruning/shortening a tree is essential to ease our understanding of the outcome and optimize it. This optimization can be done in one of three ways:

* **criterion**: optional (default="gini") or Choose attribute selection measure

Defines what function will be used to measure the quality of a split. The options are "gini" and "entropy". The Gini ratio measures the impurity of the node in a decision tree. One can assume that a node is pure when all of its records belong to the same class.

* **splitter**: string, optional (default="best") or Split Strategy

Defines the strategy used to choose the split at each node. Supported strategies are "best" to choose the best split and "random" to choose the best random split.

* **max_depth**: int or None, optional (default=None) or Maximum Depth of a Tree

Defines the maximum depth of the tree. If it’s None, the tree will be as long as possible, when all the leaves are pure (risk of overfitting the model).

```python
'''Mediante la funcion DecisionTreeClassifier generamos la estrucuta del 
arbol de decision y a traves de max_depth se puede agregar la cantidad 
que debe tener el arbol siendo un total de 3'''
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

'''A traves dela funcion fit pasaremos como pametros las variables predictoras
como la variable a predecir'''
clf = clf.fit(X_train,y_train)

'''Calculo de la prediccion con los datos obtenidos'''
y_pred = clf.predict(X_test)
```

```python
'''Evaluacion del modelo: Se obtendra el acurracy(exactitud)
del modelo a traves de Metrics perteneciente a sklearn.'''
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
'''De manera mas resumida se obtendran los datos
a traves de ka funcion classification report'''
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
'''A traves de la herramientas de pydotplus crearemos el grafico del arbol 
ya podado, indicando la funcion a graficar y se le asignara un nombre
respectivamente el cual es diabetes 2'''
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                    filled=True, rounded=True,
                    special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes2.png')
Image(graph.create_png())
```

```python
'''Posteriormente se desarollara la matriz de confucion mediante los datos
alojados en el testeo de Y y la prediccion de Y'''
Matriz_Confusion = confusion_matrix(y_test, y_pred)
Matriz_Confusion
```

```python
'''Para calcular la precision global del modelo se suman los valores correctos
predichos divididos para la suma de ltodos los valores incluidos en la matriz'''
print("Precision Global")
Precision_Global = np.sum(Matriz_Confusion.diagonal())/np.sum(Matriz_Confusion)
Precision_Global 
```

```python
'''Obtenemos la precision por cada una de las clases , ya que el modelo global
no confirma si el modelo es bueno prediciendo, de este modo se obtuvo la precision 
de no poseer diabtes '''
print("Precision por clase: No Diabetico")
Precision_No = ((Matriz_Confusion[0,0]))/sum(Matriz_Confusion[0,])
Precision_No
```

```python
'''Obtenemos la precision por cada una de las clases , ya que el modelo global
no confirma si el modelo es bueno prediciendo, de este modo se obtuvo la precision 
de si eermos '''
print("Precision por clase: Si Diabetico")
Precision_Si = ((Matriz_Confusion[1,1]))/sum(Matriz_Confusion[1,])
Precision_Si
```

```python
'''Procederemos a agregar nombres en la matriz'''

Matriz_Confusion_df = pd.DataFrame(Matriz_Confusion,
                     index = ['No diabetes', 'Si diabetes'], 
                     columns = ['No diabetes', 'Si diabetes'])
```

```python
'''Imprimira la matriz de confucion 3x3 creada a traves de 
seaborn y hetmap'''
plt.figure(figsize=(5,4))
sns.heatmap(Matriz_Confusion_df, annot=True, fmt='g')
plt.title('Matriz de confusion')
plt.ylabel('Actual')
plt.xlabel('Prediccion')
plt.show()
```
